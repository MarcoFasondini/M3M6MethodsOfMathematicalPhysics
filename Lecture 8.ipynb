## Motivation: Ideal fluid flow around an obstacle    

Understanding branch cuts and Cauchy transforms allows us to systematically solve equations involving Laplace equation. A classic example is Ideal fluid flow. Consider the case of uniform flow with angle $\theta$ around an infinitesimally small plate on $[-1,1]$. We can model this as
\begin{align*}
u(x,y) &\sim y \cos \theta - x \sin \theta  \\
u_{xx} + u_{yy} &= 0 \\
u(x,0) & = 0 \qqfor -1 < x <1
\end{align*}
Using the techniques we develop in the next few lectures, we obtain a nice, simple, closed form expression as the imaginary part of an analytic function:

```julia
μ = z -> -im*(sqrt(z-1)sqrt(z+1) - z)
Φ = (θ,z) -> exp(-im*θ)*z + sin(θ)μ(z)
u = (θ,x,y) -> imag(Φ(θ, x + im*y))
xx = yy = linspace(-3.,3.,500)
contour(xx, yy, u.(0.9,xx',yy); nlevels = 100)
plot!(Segment(-1.,1.); color=:black, label="obstacle")
```

## Algebraic powers

Before we can worry about where that solution comes from, we need to get a handle on square root functions, or mor generally, algebraic powers.  We define an algebraic power as:

**Definition (algebraic power)** 
$$z^\alpha = e^{\alpha \log z}$$

Note, for example, when $\alpha = 1/2$, $\sqrt z \equiv z^{1/2}$ is only one solution to $y^2 = z$.

```julia
α = 0.9
phaseplot(z -> z^α, (-3,3), (-3,3))
```

This has a branch cut on $(-\infty,0]$ on which it has the jump:

$${x_+^\alpha \over x_-^\alpha} = e^{\alpha (\log_+ x - \log_- x)} = e^{2 \pi \I \alpha}$$

In particular,

$$\sqrt{x}_+ = -\sqrt{x}_- = i \sqrt{|x|}$$

```julia
sqrt(-1-0.0im) , -sqrt(-1 + 0.0im)
```

These are _multiplicative jumps_. We also have a _subtractive jumps_:

\begin{align*}
x_+^\alpha - x_-^\alpha &= e^{\alpha \log_+ x} - e^{\alpha \log_- x} = e^{\alpha \log(-x) + \I  \pi \alpha} - e^{\alpha \log(-x) - \I \pi \alpha} \cr
    & = 2 \I (-x)^\alpha \sin \pi \alpha
\end{align*}

```julia
x = -1
(x+0.0im)^α - (-1-0.0im)^α, 2im*(-x)^α*sin(π*α)
```


and an _additive jump_:
\begin{align*}
x_+^\alpha + x_-^\alpha & = 2  (-x)^\alpha \cos \pi \alpha
\end{align*}

```julia
x = -1
(x+0.0im)^α + (-1-0.0im)^α, 2*(-x)^α*cos(π*α)
```

In particular, for $x < 0$,
\begin{align*}
\sqrt{x}_+ - \sqrt{x}_- &= 2 \I \sqrt{-x} \\
\sqrt{x}_+ + \sqrt{x}_- &= 0
\end{align*}

```julia
sqrt(x+0.0im) - sqrt(x-0.0im) == 2im*sqrt(-x)
```

```julia
sqrt(x+0.0im) + sqrt(x-0.0im)
```

Let's look at another example: $\varphi(z) = \sqrt{z-1}\sqrt{z+1}$. 

```julia
φ = z -> sqrt(z-1)*sqrt(z+1)
phaseplot(φ, (-3,3), (-3,3))
```

For $x > 0$ it's holomorphic. For $-1 < x < 1$ we have the multiplicative jump

$$\varphi_+(x) = \sqrt{x-1}_+ \sqrt{x+1} = - \sqrt{x-1}_- \sqrt{x+1} = -\varphi_-(x)$$
which gives the additive jump
$$
\varphi_+(x) + \varphi_-(x) = 0
$$

```julia
x = 0.1
φ(x + 0.0im) + φ(x - 0.0im)
```

 But we also have a _subtractive jump_:
$$
\varphi_+(x) - \varphi_-(x) = (\sqrt{x-1}_+ - \sqrt{x-1}_-) \sqrt{x+1} = 2\I \sqrt{1-x}\sqrt{x+1} = 2\I \sqrt{1 - x^2}
$$

```julia
x = 0.1
φ(x + 0.0im) - φ(x - 0.0im), 2im*sqrt(1-x^2)
```

For $x < -1$ we actually have continuity:

$$\varphi_+(x) = \sqrt{x-1}_+ \sqrt{x+1}_+ = (- \sqrt{x-1}_-)(- \sqrt{x+1}_-) = \varphi_-(x)$$

```julia
φ = z -> sqrt(z-1)sqrt(z+1)

x = -3.0
φ(x+0.0im), φ(x-0.0im)
```

# Inferred analyticity

But continuity _implies_ analyticity, using Cauchy's integral formula:

**Theorem (continuity on a curve implies analyticity)** Let $D$ be a domain and $\gamma \subset D$ a contour. Suppose $f$ is analytic in $D \backslash \gamma$, and continuous on the interior of $\gamma$. Then $f$ is analytic in $D \backslash \{\gamma(a), \gamma(b) \}$.

**Theorem (weaker than pole singularity implies analyticity)** Suppose $f$ is analytic in $D \backslash \{ z_0 \}$ and has a weaker than pole singularity at $z_0$:
$$\lim_{z \rightarrow z_0} (z-z_0) f(z) = 0$$
holds uniformly. Then $f$ is analytic at $z_0$. (More precisely: $f$ can be analytically continued to $z_0$.)


**Corollary (Weaker than linear growth implies analyticity at infinity)** If 
$$
\lim_{z \rightarrow \infty} {f(z)  \over z} = 0,
$$ then $f$ is analytic at infinity.


From these result we can infer that 
$$\phi(z) = \sqrt{z-1}\sqrt{z+1}$$
is analytic on $\C \backslash [-1,1]$, and $\phi(z) \sim_{z \rightarrow \infty} z$.


### Uniqueness results from inferred singularity

Recall that an important ingredient of complex analysis is Liouville's theorem:

**Theorem (Liouville)** If $f$ is entire and  bounded in ${\mathbb C}$, then $f$ must be constant.


We will see that knowledge of the behaviour of $\phi$ can be used to recover by it's behaviour at it's singularities at $\infty$ and jump on $[-1,1]$. But before that, we already have the following uniqueness results by combining the above with Liouville's theorem:


1. $\phi(z)$ is the unique function analytic in $\C \backslash [-1,1]$ with weaker than pole singularities at $\pm 1$ satisfying $\phi(z) \sim z$ and 
$$\phi_+(x) - \phi_-(x) = 2\I \sqrt{1-x^2} \qqfor -1 < x <1.$$ 
2. $\kappa(z) = {1 \over \sqrt{z-1} \sqrt{z+1}}$ is the unique function analytic in $\bar\C \backslash [-1,1]$ with weaker than pole singularities at $\pm 1$ satisfying $\kappa(\infty) = 0$ and 
$$
\kappa_+(x) - \kappa_-(x) = {-2\I \over \sqrt{1-x^2}}\qqfor -1 < x <1.
$$


```julia
κ = z -> 1/(sqrt(z-1)sqrt(z+1))
phaseplot(κ, (-3,3), (-3,3))
```

```julia
x = 0.1
κ(x + 0.0im) - κ(x - 0.0im) , 2im/sqrt(1-x^2)
```

```julia
x = -2.3
κ(x + 0.0im) - κ(x - 0.0im) 
```

3\. $\mu(z) = \log(z +1) - \log(z-1)$ is the unique function analytic in $\C \backslash [-1,1]$ with weaker than pole singularities at $\pm 1$ satisfying $\mu(\infty) = 0$ and 
$$\mu_+(x) - \mu_-(x) = 2 \pi \I \qqfor -1 < x < 1.$$



```julia
μ = z -> log(z+1) - log(z-1)
phaseplot(μ, (-3,3), (-3,3))
```

```julia
x = 0.3
μ(x + 0.0im) - μ(x - 0.0im) 
```

# Cauchy transform on the interval

We can consistently construct a function with a prescribed subtractive jump using the _Cauchy transform_, which really is just Cauchy's integral formula but where we know nothing about $f(z)$:

**Definition (Cauchy transform)** 
$$
\CC_\gamma f(z) := {1 \over 2 \pi \I} \int_\gamma {f(\zeta) \over \zeta - z} \D\zeta
$$


We focus on the case of an interval $[a,b]$:
$$
\CC_{[a,b]} f(z) := {1 \over 2 \pi \I} \int_a^b {f(x) \over x - z} \dx
$$

```julia
x = Fun(-1 .. 1)
f = exp(x)*sqrt(1-x^2)
phaseplot(z -> cauchy(f,z), (-3,3),(-3,3))
```

```julia
cauchy(f, 2+2im ), sum(f/(x-(2+2im)))/(2π*im)
```

It turns out that the Cauchy transform has a very simple subtractive jump: 

**Theorem (Plemelj on the interval I)**
Suppose $f$ is differentiable on $(a,b)$ and is absolutely integrable on $[a,b]$. Then the Cauchy transform has the following properties:
1. $\CC_{[a,b]} f(z)$ is analytic in $\bar \C \backslash [a,b]$
2. $\CC_{[a,b]} f(z)$ has weaker than pole singularities at $a$ and $b$
3. $\CC_{[a,b]} f(\infty) = 0$
4. It has the subtractive jump:
$$
\CC_{[a,b]}^+ f(x) - \CC_{[a,b]}^- f(x) = f(x) \qqfor a < x < b
$$

```julia
cauchy(f, 0.1+0.0im)-cauchy(f, 0.1-0.0im) , f(0.1)
```

We can use the results of the previous results to show that it is in fact unique.


**Theorem (Plemelj on the interval II)** Suppose $\phi(z)$ satsfies the following properties:
1. $\phi(z)$ is analytic in $\bar \C \backslash [a,b]$
2. $\phi(z)$ has weaker than pole singularities at $a$ and $b$
3. $\phi(\infty) = 0$
4. It has the subtractive jump:
$$
\phi^+(x) - \phi^-(x) = f(x) \qqfor a < x < b
$$
where $f$ is differentiable in $(a,b)$ and absolutely integrable in $[a,b]$. 

Then $\phi(z) = {\cal C}_{[a,b]} f(z)$.

_Examples_ We can use this theorem to prove the following relationships:

$$
{1 \over \sqrt{z-1} \sqrt{z+1}} = -2 \I {\cal C}\left[{1 \over \sqrt{1-x^2}}\right](z) = -{1 \over \pi}\int_{-1}^1 {\dx \over \sqrt{1-x^2} (x-z)}
$$

```julia
z = 2 +2im
1/(sqrt(z-1)sqrt(z+1)),-2im*cauchy(1/sqrt(1-x^2),z)
```

We can subtract out the singularity at infinity first to determine

$$\sqrt{z-1}\sqrt{z+1} = z  + 2 \I {\cal C}[\sqrt{1-x^2}](z)$$

```julia
sqrt(z-1)sqrt(z+1), z +2im*cauchy(sqrt(1-x^2),z)
```

Finally, we have the following (also verifiable using indefinite integration):

$$\log(z+1) - \log(z-1) = 2 \pi \I {\cal C}[1](z) = \int_{-1}^1 {\dx \over x -z} $$

## Hilbert transform on the interval

We now investigate the additive jump that the Cauchy transform satisfies.

**Definition (Principal value integral on an interval)** 
The (Cauchy) principal value integral at $x_0$ is defined as
$$
\dashint_a^b f(x) \dx := \lim_{\epsilon \rightarrow 0} \left(\int_a^{x_0 - \epsilon} + \int_{x_0 + \epsilon}^b \right) f(x) dx
$$


_Examples_:

1. $$\dashint_{-1}^1  {1 \over x} \dx = 0$$
2. $$\dashint_{-1}^1  \cot x \dx = 0$$
2. $$\dashint_{-1}^1  {1 \over t - x} \D t = \log(1+x) - \log(1-x)$$

```julia
ε = 0.01
x = 0.1
Γ = Segment(-1 , x-ε) ∪ Segment(x+ε , 1)
t = Fun(Γ)
f = Fun(1/(t - x))
plot(f; legend=false, title="integral = $(sum(f))",ylims=(-20,20))
```

```julia
log(1-x) - log(1+x)
```

**Definition (Hilbert transform)** For $a < x < b$, define
$$
\HH_{[a,b]} f(x) =  {1 \over \pi} \dashint_a^b {f(t) \over {t - x}} \D t
$$

```julia
x = Fun(-1 .. 1)
f = exp(x)*sqrt(1-x^2)
plot(f; title="function")
plot!(hilbert(f); title="hilbert transform")
```


**Theorem (Plemelj on the interval III)** 
The Cauchy transform has the additive jump
$$
\CC_{[a,b]}^+ f(x) + \CC_{[a,b]}^- f(x) = - \I \HH f(x)
$$

```julia
cauchy(f, 0.1+0.0im)+cauchy(f, 0.1-0.0im) , -im*hilbert(f)(0.1)
```
